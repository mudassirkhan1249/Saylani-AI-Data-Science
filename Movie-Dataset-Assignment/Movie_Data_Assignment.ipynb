### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
def assign_genre_group(genre_str):
    if not isinstance(genre_str, str):
        return None
    has_action = 'Action' in genre_str
    has_drama = 'Drama' in genre_str
    
    if has_action and not has_drama:
        return 'Action'
    elif has_drama and not has_action:
        return 'Drama'
    else:
        return None
df['genre_group'] = df['Genre'].apply(assign_genre_group)

df_plot = df.dropna(subset=['genre_group'])

print(f"Action movies: {len(df_plot[df_plot['genre_group'] == 'Action'])}")
print(f"Drama movies: {len(df_plot[df_plot['genre_group'] == 'Drama'])}")

plt.figure(figsize=(12, 6))

sns.boxplot(x='genre_group', y='Gross', data=df_plot, palette='Set2')
plt.yscale('log')  

plt.title('Comparison of Gross Revenue: Action vs Drama Movies', fontsize=16)
plt.xlabel('Genre Group', fontsize=12)
plt.ylabel('Gross Revenue (USD) - Log Scale', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
df['IMDB_Rating_scaled'] = df['IMDB_Rating'] * 10

top5 = df.nlargest(5, 'IMDB_Rating')

movies = top5['Series_Title'].values
meta = top5['Meta_score'].values
imdb_scaled = top5['IMDB_Rating_scaled'].values

x = np.arange(len(movies))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, meta, width, label='Meta_score')
bars2 = ax.bar(x + width/2, imdb_scaled, width, label='IMDB_Rating (scaled to 100)')

ax.set_xlabel('Movie')
ax.set_ylabel('Score')
ax.set_title('Top 5 Highest Rated Movies: Meta_score vs IMDB_Rating (scaled)')
ax.set_xticks(x)
ax.set_xticklabels(movies, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
