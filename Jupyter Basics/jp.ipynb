- Mathematics
- Statistics

In simple words:

> Data Science = Raw Data → Analysis → Insights → Better Decisions

# What is Data Science?

Data Science is the process of collecting, analyzing, and interpreting large amounts of data to discover meaningful patterns.

The main goal is to transform raw data into useful knowledge.

Mathematically we can represent it as:

$$
Knowledge = Data + Analysis
$$

# Why Data Science is Important?

In the modern digital world, massive amounts of data are generated every second.

Examples of data sources include:

- Social Media Platforms
- Online Shopping Websites
- Banking Systems
- Healthcare Records

Companies use Data Science to:

1. Make better business decisions
2. Understand customer behavior
3. Predict future trends

# Data Science Workflow

The typical Data Science process consists of several steps:

1. Data Collection
2. Data Cleaning
3. Data Analysis
4. Data Visualization
5. Model Building
6. Prediction and Deployment

A simple predictive model can be written as:

$$
y = f(x)
$$

Where:

- $x$ = input data
- $y$ = predicted output
