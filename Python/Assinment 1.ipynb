# 🔹 Q12 (Full Name + Upper/Lower)  User se first name aur last name lo aur:  1)  full name print karo  2) upper case 3)lower case

```python
# type your code here
```

```python
first_name = input("Enter your first name :")
second_name = input("Enter your second name :")
full_name = first_name + " " + second_name
print(full_name)
print(full_name.upper())
print(full_name.lower())
```

# 🔹 Q15 (Simple Bank Info / ID Card Style ) User se: 1) name  2)CNIC  3)balance lo aur formatted output dikhao

```python
name = input("Enter your name: ")
cnic = input("Enter your CNIC (without dashes or with dashes): ")
balance = float(input("Enter your account balance: "))

print("\n" + "=" * 40)
print("         SIMPLE BANK INFO / ID CARD")
print("=" * 40)
print(f"Name    : {name.title()}")
print(f"CNIC    : {cnic}")
print(f"Balance : PKR {balance}")
print("=" * 40)
```
