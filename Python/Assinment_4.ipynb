**Q2: `while loop` ka istemal karte hue 10 se 1 tak reverse counting print karein.**

```python
num = 10
x = True
while x:
    if num >= 0:
        print(num, end = " ")
        num -= 1
    if num == 0:
        x = False
```

**Q12: Check karein ke number Prime hai ya nahi.**

```python
def check_prime(number):
    if number <= 1:
        return False
        
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False 
            
    return True 

num = int(input('enter a number'.title()))
if check_prime(num):
    print(f"{num} is Prime Number.")
else:
    print(f"{num} is not Prime Number.")
```

**Q20: Number Guessing Game (3 chances).**

```python
import random

secret_number = random.randint(1, 10)
chances = 3

print("Welcome to the Number Guessing Game!")
print("I am thinking of a number between 1 and 10. You have 3 chances to guess it.")

for attempt in range(1, chances + 1):
    guess = int(input(f"Attempt {attempt}/{chances}: Enter your guess: "))
    
    if guess == secret_number:
        print(f"Congratulations! You guessed it right. The number was {secret_number}! 🎉")
        break  
    elif guess < secret_number:
        print("Too low! Try a higher number. 📈")
    else:
        print("Too high! Try a lower number. 📉")

else:
    print(f"\nGame Over! You ran out of chances. The correct number was {secret_number}. 😢")
```

#### **Q29: Recursive function banayein jo factorial calculate kare.**

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

factorial(5), factorial(0), factorial(1)
```

#### *Q30: Function banayein jo user se unlimited numbers le (*args) aur sirf even numbers ki list return kare.**

```python
def filtering_even(*args):
    even = []
    for i in args:
        if i % 2 == 0:
            even.append(i)

    return even

filtering_even(1,2,3,4,5,6,7,8,9,0), filtering_even(11,13,15,17,19)
```
