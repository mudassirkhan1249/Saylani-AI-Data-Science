#### Q10: Count Vowels in String

```python
st = 'mudassir'
vowels = 'aeiou'
count = 0
for i in st:
    for j in vowels:
        if j in i:
            count += 1

print(count)
```

#### Q11: Reverse List using Loop

```python
lst = [1, 2, 3, 4, 5]
lst2 = []

for i in range(len(lst) - 1, -1, -1):
    lst2.append(lst[i])

print(lst2)
```
