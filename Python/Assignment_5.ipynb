**Q1:** Aik `Student` class banayein jismein `name` aur `roll_number` private variables (`__`) hon. In variables ko access aur modify karne ke liye getter aur setter methods likhein.

```python
class Student:
    def __init__(self, name, roll_number):
        self.__name = name
        self.__roll_number = roll_number

    def get_name(self):
        return self.__name

    def get_roll_number(self):
        return self.__roll_number

    def set_name(self, new_name):
        if isinstance(new_name, str) and len(new_name.strip()) > 0:
            self.__name = new_name
        else:
            print("Error: Invalid name!")

    def set_roll_number(self, new_roll):
        if isinstance(new_roll, int) and new_roll > 0: 
            self.__roll_number = new_roll
        else:
            print("Error: Roll number must be a positive integer!")
```

**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}. New Balance: {self.__balance}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Error: Insufficient balance! Transaction denied.")
        elif amount <= 0:
            print("Invalid withdrawal amount!")
        else:
            self.__balance -= amount
            print(f"Withdrawn: {amount}. Remaining Balance: {self.__balance}")

# Test code
account = BankAccount(1000)
account.deposit(500)
account.withdraw(2000)
account.withdraw(400)
```

**Q3:** Aik class banayein `Employee` jismein salary private ho. Agar koi salary ko minus (negative) value set karne ki koshish kare, toh setter method 'Invalid Salary' ka error message show kare.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = 0
        self.set_salary(salary)

    def get_salary(self):
        return self.__salary

    def set_salary(self, new_salary):
        if new_salary >= 0:
            self.__salary = new_salary
        else:
            print("Error: Invalid Salary! Salary cannot be negative.")

# Test code
emp = Employee("Kamran", 50000)
print(f"Salary: {emp.get_salary()}")
emp.set_salary(-1500)
```

**Q7:** Multilevel Inheritance ka use karein. Class `Animal` -> Class `Mammal` -> Class `Dog`. Dog class ke object se parent classes ke features ko access karein.

```python
class Animal:
    def eat(self):
        print("This animal eats food.")

class Mammal(Animal):
    def walk(self):
        print("This mammal walks on legs.")

class Dog(Mammal):
    def bark(self):
        print("The dog barks: Woof! Woof!")

my_dog = Dog()
my_dog.eat()
my_dog.walk()
my_dog.bark()
```

**Q9:** Child class ke constructor (`__init__`) ke andar parent class ke constructor ko call karne ke liye `super()` ka use kaise hota hai? Aik practical example likhein.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
class Specialist(Person):
    def __init__(self, name, age, specialization):
        super().__init__(name, age)
        self.specialization = specialization

doc = Specialist("Dr. Asif", 45, "Cardiologist")
print(f"Name: {doc.name}, Age: {doc.age}, Field: {doc.specialization}")
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

**Q11:** Method Overriding kya hai? Parent class `Bird` banayein jisme `fly()` method ho jo print kare 'Most birds can fly'. Child class `Ostrich` banayein jo is method ko override kare aur print kare 'Ostriches cannot fly'.

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Book:
    def __init__(self, pages):
        self.pages = pages

    def __add__(self, other):
        return self.pages + other.pages

book1 = Book(250)
book2 = Book(150)
total_pages = book1 + book2
print(f"Total Pages combined: {total_pages}")
```

**Q16:** `abc` module ka use karte hue aik abstract class `Computer` banayein jisme abstract method `process()` ho. Is class ka object banane ki koshish karein aur dekhein kya hota hai.

```python
from abc import ABC, abstractmethod

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass

try:
    my_comp = Computer()
except TypeError as e:
    print(f"Caught Expected Error: {e}")
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    def start_engine(self):
        print("Engine started... Vroom Vroom!")

    @abstractmethod
    def drive(self):
        pass

class ElectricCar(Car):
    def drive(self):
        print("Driving smoothly without noise.")

my_ev = ElectricCar()
my_ev.start_engine()
my_ev.drive()
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    def __init__(self, amount):
        self.amount = amount

    def print_receipt(self):
        print(f"Receipt: Payment of Rs.{self.amount} successful.")

    @abstractmethod
    def process_payment(self):
        pass

class CreditCardPayment(Payment):
    def __init__(self, amount, card_num):
        super().__init__(amount)
        self.card_num = card_num

    def process_payment(self):
        print(f"Processing Credit Card payment of Rs.{self.amount} via Card: ****{self.card_num[-4:]}")

class UPIPayment(Payment):
    def __init__(self, amount, upi_id):
        super().__init__(amount)
        self.upi_id = upi_id

    def process_payment(self):
        print(f"Processing UPI payment of Rs.{self.amount} via ID: {self.upi_id}")

payment1 = CreditCardPayment(5000, "1234567890123456")
payment1.process_payment()
payment1.print_receipt()

print("---")

payment2 = UPIPayment(1200, "user@upi")
payment2.process_payment()
payment2.print_receipt()
```
